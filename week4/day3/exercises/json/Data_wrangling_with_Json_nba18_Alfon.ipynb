## Working on Jsons

```python
import requests
import json
import pandas as pd
```

```python
r = requests.get(url='http://data.nba.net/prod/v2/2018/teams.json')
json_readed = json.loads(r.text)
json_readed
```

```python
json_df = pd.DataFrame(json_readed)
json_df
```

### Data Mining & Data Wrangling
As you can see, there are jsons inside the original json. For that, we have to mining the json (data mining) to be able to use it correctly (data wrangling). 

How do you solve this issue? Research about this and try a solution. 

```python
league = pd.DataFrame(json_df['league'])
league
```

```python
idx = pd.Index(['standard'])
a = pd.DataFrame(league.loc['standard', 'league'], index=idx.repeat(len(a)))
a
```

```python
idx = pd.Index(['africa'])
b = pd.DataFrame(league.loc['africa', 'league'], index=idx.repeat(len(b)))
b
```

```python
idx = pd.Index(['sacramento'])
c = pd.DataFrame(league.loc['sacramento', 'league'], index=idx.repeat(len(c)))
c
```

```python
idx = pd.Index(['vegas'])
d = pd.DataFrame(league.loc['vegas', 'league'], index=idx.repeat(len(d)))
d
```

```python
idx = pd.Index(['utah'])
e = pd.DataFrame(league.loc['utah', 'league'], index=idx.repeat(len(e)))
e
```

```python
together = pd.concat([a, b, c, d, e])
together
```
