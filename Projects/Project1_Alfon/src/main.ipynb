## Calcular Umbrales de manera automatica

```python
low_thres = float("{:.2f}".format(max(df['radius_mean'])))
high_thres = float("{:.2f}".format(min(df['radius_mean'])))
ben_x_pcn = 0
mal_x_pcn = 0
while ben_x_pcn < 95:
    low_thres -= 0.01
    df_x = df[df['radius_mean'] < low_thres]
    tot_x = len(df_x)
    ben_x = df_x[df_x['diagnosis'] == 'B']
    ben_x_pcn = len(ben_x) / tot_x * 100
low_thres = round(low_thres, 2)
while mal_x_pcn < 95:
    high_thres += 0.01
    df_x = df[df['radius_mean'] > high_thres]
    tot_x = len(df_x)
    mal_x = df_x[df_x['diagnosis'] == 'M']
    mal_x_pcn = len(mal_x) / tot_x * 100
high_thres = round(high_thres, 2)
```

## Mostrar Porcentajes de malignidad y beningnidad a partir de los umbrales anteriores

```python
def malignant_percentage_over_thres (x):
    df_x = df[df['radius_mean'] > x]
    tot_x = len(df_x)
    ben_x = df_x[df_x['diagnosis'] == 'B']
    mal_x = df_x[df_x['diagnosis'] == 'M']
    ben_x_pcn = len(ben_x) / tot_x * 100
    mal_x_pcn = len(mal_x) / tot_x * 100
    plt.pie([ben_x_pcn, mal_x_pcn], labels=('Benign', 'Malignant'), autopct='%1.1f%%')
    plt.title('Diagnosis Distribution at Radius > ' + str(x) + 'mm')
    return round(ben_x_pcn, 1), round(mal_x_pcn, 1)
```

```python
malignant_percentage_over_thres(high_thres)
```

```python
def malignant_percentage_under_thres (x):
    df_x = df[df['radius_mean'] < x]
    tot_x = len(df_x)
    ben_x = df_x[df_x['diagnosis'] == 'B']
    mal_x = df_x[df_x['diagnosis'] == 'M']
    ben_x_pcn = len(ben_x) / tot_x * 100
    mal_x_pcn = len(mal_x) / tot_x * 100
    plt.pie([ben_x_pcn, mal_x_pcn], labels=('Benign', 'Malignant'), autopct='%1.1f%%')
    plt.title('Diagnosis Distribution at Radius < ' + str(x) + 'mm')
    return round(ben_x_pcn, 1), round(mal_x_pcn, 1)
```

```python
malignant_percentage_under_thres(low_thres)
```

```python
def malignant_percentage_between_thres (x, y):
    df_x = df[(df['radius_mean'] >= x) & (df['radius_mean'] <= y)]
    tot_x = len(df_x)
    ben_x = df_x[df_x['diagnosis'] == 'B']
    mal_x = df_x[df_x['diagnosis'] == 'M']
    ben_x_pcn = len(ben_x) / tot_x * 100
    mal_x_pcn = len(mal_x) / tot_x * 100
    plt.pie([ben_x_pcn, mal_x_pcn], labels=('Benign', 'Malignant'), autopct='%1.1f%%')
    plt.title('Diagnosis Distribution at ' + str(x) +  ' ≤ R ≤ '  + str(y) + 'mm')
    return round(ben_x_pcn, 1), round(mal_x_pcn, 1)
```

```python
malignant_percentage_between_thres (low_thres, high_thres)
```

## Distribuir a los pacientes en grupos y crear un DF

```python
the_list = []
ind = []
x = low_thres
df_x = df[df['radius_mean'] < x]
tot_x = len(df_x)
ben_x = df_x[df_x['diagnosis'] == 'B']
mal_x = df_x[df_x['diagnosis'] == 'M']
ben_x_pcn = len(ben_x) / len(df_ben) * 100
mal_x_pcn = len(mal_x) / len(df_mal) * 100
tot_x_pcn = tot_x / len(df) * 100
the_list.append([len(ben_x), round(ben_x_pcn, 2), len(mal_x), round(mal_x_pcn, 2), tot_x, round(tot_x_pcn, 2)])
ind.append('R < ' + str(x))
```

```python
x, y = low_thres, high_thres
df_x = df[(df['radius_mean'] >= x) & (df['radius_mean'] <= y)]
tot_x = len(df_x)
ben_x = df_x[df_x['diagnosis'] == 'B']
mal_x = df_x[df_x['diagnosis'] == 'M']
ben_x_pcn = len(ben_x) / len(df_ben) * 100
mal_x_pcn = len(mal_x) / len(df_mal) * 100
tot_x_pcn = tot_x / len(df) * 100
the_list.append([len(ben_x), round(ben_x_pcn, 2), len(mal_x), round(mal_x_pcn, 2), tot_x, round(tot_x_pcn, 2)])
ind.append(str(x) + ' < R < ' + str(y))
```

```python
x = high_thres
df_x = df[df['radius_mean'] > x]
tot_x = len(df_x)
ben_x = df_x[df_x['diagnosis'] == 'B']
mal_x = df_x[df_x['diagnosis'] == 'M']
ben_x_pcn = len(ben_x) / len(df_ben) * 100
mal_x_pcn = len(mal_x) / len(df_mal) * 100
tot_x_pcn = tot_x / len(df) * 100
the_list.append([len(ben_x), round(ben_x_pcn, 2), len(mal_x), round(mal_x_pcn, 2), tot_x, round(tot_x_pcn, 2)])
ind.append('R > ' + str(x))
```
